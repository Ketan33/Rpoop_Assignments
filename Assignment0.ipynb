```python
#rectangle functions
class Rectangle:
    def __init__(self):
        self.length=0
        self.breadth=0
        
    def set_length(self,length):
        self.length = length
        
    def set_breadth(self,breadth):
        self.breadth = breadth
        
    def area(self):
        print(f"Area is {self.length*self.breadth} units square")
        
    def perimeter(self):
        print(f"Perimeter is {2*(self.length+self.breadth)} units")
        
    def get_length(self):
        print(f"Length={self.length} units")
        
    def get_breadth(self):
        print(f"Breadth={self.breadth}")
```

```python
rect = Rectangle()
while True:
    print("1.Set the length")
    print("2. Set the breadth")
    print("3. Get the length")
    print("4. Get the breadth")
    print("5. Get area")
    print("6. Get perimeter")
    print("7.Exit")
    
    choice = int(input("Enter the choice:"))
    
    if choice==1:
        length = float(input("Enter the length:"))
        rect.set_length(length)
        
    elif choice==2:
        breadth = float(input("Enter the breadth:"))
        rect.set_breadth(breadth)
        
    elif choice==3:
        rect.get_length()
        
    elif choice==4:
        rect.get_breadth()
        
    elif choice==5:
        rect.area()

    elif choice==6:
        rect.perimeter()
        
    elif choice==7:
        break
        
    else :
        print("Invalid input")
```
