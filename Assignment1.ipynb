```python
#Design and implement classes representing a polygon shape, and derived classes like rectangle, 
#and square with the same operations as mentioned in assignment above. Write code to test your classes.
import math

class Polygon:
    def __init__(self,number_sides,length_sides):
        self.number_sides = number_sides
        self.length_sides = length_sides
    
    def perimeter(self):
        return sum(self.length_sides)
    
    def area(self):
        pass
    
class Rectangle(Polygon):
    def __init__(self,length,breadth):
        super().__init__(4,[length,breadth,length,breadth])
        self.length = length
        self.breadth = breadth
        
    def area(self):
        return self.length*self.breadth
    
class Square(Rectangle):
    def __init__(self,side):
        super().__init__(side,side)
        self.side = side
        
class Triangle(Polygon):
    def __init__(self,side1,side2,side3):
        super().__init__(3,[side1,side2,side3])
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3
        
    def area(self):
        s = self.perimeter/2
        return math.sqrt(s*(s-self.side1)*(s-self.side2)*(s-self.side3))
    
while True:
    print("1. Polygon")
    print("2. Rectangle")
    print("3. Square")
    print("4. Triangle")
    print("5. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        n = int(input("Number of sides of polygon:"))
        lst =[]
        for i in range(n):
            try:
                lst.append(int(input("Enter side:")))
            except:
                print("Input Error")
                break
                
        polygon = Polygon(n, lst)
        print("Polygon Perimeter", polygon.perimeter())
    elif choice == 2:
        try:
            rectangle = Rectangle(int(input("Enter side 1:")), int(input("Enter side 2:")))
            print("Rectangle Perimeter",rectangle.perimeter())
            print("Rectangle Area",rectangle.area())
        except:
            print("Input Error")
    elif choice == 3:
        try:     
            square = Square(int(input("Enter side:")))
            print("Square Perimeter",square.perimeter())
            print("Square Area",square.area())
        except:
            print("Input Error")
    elif choice == 4:
        try:
            triangle = Triangle(int(input("Enter side 1:")),int(input("Enter side 2:")),int(input("Enter side 3:")))
            print(triangle.area())
        except:
            print("Input Error")
    elif choice == 5:
        break
    else:
        print("Invalid choice. Please try again.")
```
