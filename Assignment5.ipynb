```python
import tkinter as tk

def label1_clicked():
    """Update the text of label1"""
    label1.config(text="Hello world", bg="orange", fg="white", font=("Arial",20,"bold"))
    
def label2_hover(event):
    """Change the background when mouse hovers"""
    label2.config(bg="#88cffa",bd=5)
    
def label2_unhover(event):
    """Change the background color and border of label 2 when the mouse leaves"""
    label2.config(bg="#ffd166",bd=1)
#create window
root = tk.Tk()
root.title("Colorful labels")

#set default font and background color
root.option_add("*Font", "Arial 14")
root.config(bg="#282c34")

#create two labels with different backgrounds
label1 = tk.Label(root, text="Label 1", bg="#ff7f50", fg="#282c34", width=20, height=5, bd=3, relief="groove")
label1.grid(row=0,column=0)

label2 = tk.Label(root,text="Label 2",bg="#9fe2bf",fg="#282c34",width=20, height=5, bd=1, relief="solid")
label2.grid(row=1,column=0)

#on and off click
label1.bind("<Button-1>", lambda event:label1_clicked())
label2.bind("<Enter>",label2_hover)
label2.bind("<Leave>",label2_unhover)

root.mainloop()
```

```python
import tkinter as tk
root = tk.Tk()
root.geometry("300x300")
my_menu = tk.Menu(root)
root.config(menu=my_menu)

#click command
def function():
    my_label = tk.Label(root,text="Clicked a dropdown menu").pack()

def file_new():
    hide_all_frames()
    file_new_frame.pack(fill="both",expand=1)
    my_label = tk.Label(root,text="Clicked a file menu").pack()
    
def edit_cut():
    hide_all_frames()
    edit_cut_frame.pack(fill="both",expand=1)
    my_label = tk.Label(root,text="Clicked a edit menu").pack()
    
def hide_all_frames():
    file_new_frame.pack_forget()
    edit_cut_frame.pack_forget()
    
#create a menu item
file_menu = tk.Menu(my_menu,tearoff=0)
my_menu.add_cascade(label="File", menu=file_menu)
file_menu.add_command(label="New", command=function)
file_menu.add_command(label="Open", command=file_new)
file_menu.add_command(label="Save", command=function)
file_menu.add_separator()
file_menu.add_command(label="Close", command=root.quit)

#create an edit menu item
edit_menu = tk.Menu(my_menu,tearoff=0)
my_menu.add_cascade(label="Edit",menu=edit_menu)
edit_menu.add_command(label="Cut", command = edit_cut)
edit_menu.add_command(label="Copy", command = function)
edit_menu.add_command(label="Paste", command = function)

#create options menu
options_menu = tk.Menu(my_menu,tearoff=0)
my_menu.add_cascade(label="Options", menu=options_menu)
options_menu.add_command(label="Find", command = function)
options_menu.add_command(label="Find Next", command = function)

#adding frames to demonstrate functionality
file_new_frame = tk.Frame(root,width=300, height=300, bg="red")
edit_cut_frame = tk.Frame(root,width=300, height=300, bg="blue")

root.mainloop()
```
