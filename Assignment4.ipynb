```python
import random
class BankAccount:
    def __init__(self,name,acc_no,acc_type,balance):
        self.name = name
        self.acc_no = acc_no
        self.acc_type = acc_type
        self.balance = balance
        
    def withdraw(self,amount):
        if (amount>self.balance):
            print("Insufficient balance")
        else :
            self.balance-=amount
            print(f"Withdrawn {amount}.New balance is {self.balance}")
            
    def display_detail(self):
        print(f"Name:{self.name}")
        print(f"Account number:{self.acc_no}")
        print(f"Account type:{self.acc_type}")
        print(f"Balance:{self.balance}")
        
    def deposit(self,amount):
        self.balance+=amount
        print(f"Deposited {amount}.New balance is {self.balance}")
        
accounts = {}

def create_account():
    try:
        name = input("Enter your name:")
        acc_no = str(random.randint(1000000,9000000))
        print("Your account number is ",acc_no)
        acc_type = input("Enter account type:")
        balance = int(input("Enter initial balance:"))
        accounts[acc_no] = BankAccount(name,acc_no,acc_type,balance)
        print("Account created successfully")
    except:
        print("Input error")
        
def deposit_money():
    try:
        acc_no = input("Enter account number:")
        amount = int(input("Enter money to deposit:"))
        accounts[acc_no].deposit(amount)
    except:
        print("Input error")
        
def withdraw_money():
    acc_no = input("Enter account number:")
    amount = int(input("Enter money to withdraw:"))
    accounts[acc_no].withdraw(amount)
    

def display_details():
    acc_no = input("Enter account number:")
    accounts[acc_no].display_detail()
    
while True:
    print("1. Create account")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Display details")
    print("5. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        create_account()
    elif choice == 2:
        deposit_money()
    elif choice == 3:
        withdraw_money()
    elif choice == 4:
        display_details()
    elif choice == 5:
        break
    else:
        print("Invalid choice. Please try again.")
        
```
