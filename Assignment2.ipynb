```python
class Vehicle:
    def __init__(self,brand):
        self.brand = brand
        
    def drive(self):
        print(f"Driving a {self.brand} vehicle")
        
class Car(Vehicle):
    def __init__(self,brand,model):
        super().__init__(brand)
        self.model = model
        
    def honk(self):
        print(f"{self.brand} {self.model} is honking")
        
class Sedan(Car):
    def __init__(self,brand,model,color):
        super().__init__(brand,model)
        self.color = color
    
    def get_color(self):
        print(f"{self.brand} {self.model} has color {self.color}")
        
class SUV(Car):
    def __init__(self,brand,model,color):
        super().__init__(brand,model)
        self.color = color
        
    def get_color(self):
        print(f"{self.brand} {self.model} has color {self.color}")
        
class Owner:
    def __init__(self,name,age):
        self.name = name
        self.age = age
        
class SUV_Owner(Owner,SUV):
    def __init__(self,name,age,brand,model,color):
        Owner.__init__(self,name,age)
        SUV.__init__(self,brand,model,color)
        
while True:
    print("1. Single Inheritence")
    print("2. Multi-level Inheritence")
    print("3. Hierarchial Inheritence")
    print("4. Multiple Inheritence")
    print("5. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        car = Car(input("Enter Car Brand"), input("Enter Car Model"))
        car.honk()
        car.drive()
    elif choice == 2:
        sedan = Sedan(input("Enter Sedan Car Brand"), input("Enter Sedan Car Model"), input("Enter Sedan Car Color"))
        sedan.drive()
        sedan.honk()   
        sedan.get_color()
    elif choice == 3:
        hatchback = SUV(input("Enter SUV Car Brand"), input("Enter SUV Car Model"), input("Enter SUV Car Color"))
        hatchback.drive()
        hatchback.honk()   
        hatchback.get_color()
    elif choice == 4:
        sedanOwner = SUV_Owner(input("Enter Owner Name"), input("Enter Owner Age"), input("Enter Hatchback Car Brand"), input("Enter Hatchback Car Model"), input("Enter Hatchback Car Color"))
        
        sedanOwner.drive()
        sedanOwner.honk()   
        sedanOwner.get_color()
    
    elif choice == 5:
        break
    else:
        print("Invalid choice. Please try again.")
```
